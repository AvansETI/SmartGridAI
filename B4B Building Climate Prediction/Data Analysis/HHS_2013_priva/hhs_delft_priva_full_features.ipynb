## Transformation

```python
import pandas as pd
import re
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

#### Load dataset

```python
df = pd.read_csv('../data/2013_HHS_Delft_priva_16min.csv', sep=',')
# df = pd.read_excel('../data/2013_HHS_Delft_priva_16min.xlsx')
```

```python
df.head()
```

### Filter out columns that have a higher amount of NaN values than normal.

```python
# In general, columns have default a NaN count of 745
# Check columns that contain a NaN sum higher than 745

max_count = 745

eliminated_columns = []

for column in df:
    col_index = df.columns.get_loc(column)
    col_nan_count = df[column].isna().sum()

    if (col_nan_count > 745):
        eliminated_columns.append(col_index)
        print(col_index, " | ", column, " | ", col_nan_count)
```

```python
eliminated_columns
```

Drop the eliminated columns

```python
dft = df.drop(df.columns[eliminated_columns], axis=1)
```

```python
dft.head()
```

```python
# Drop 'Unnamed 0' column
dft = dft.drop(df.columns[0], axis=1)
```

```python
dft.head()
```

Drop rows that contain any NaN values

```python
dft.dropna(inplace=True)
```

```python
dft.isna().sum()
```

Create 'Timestamp' column for timeseries forecasting

```python
dft['Timestamp'] = dft['Datum (d/M/yyyy)'] + ' ' + dft['Tijd']
```

```python
# Move last column to first

cols = list(dft.columns)
cols = [cols[-1]] + cols[:-1]
dft = dft[cols]
```

```python
dft = dft.drop(columns=['Datum (d/M/yyyy)', 'Tijd'])
```

```python
dft.columns.values
```

```python
# Amount of columns
len(dft.columns)
```

### Apply column name formatting
This is done so that the columns can be used properly in the configurations of the framework solution (climate-prediction).

```python
dft_clean = dft.copy()

# Previous column name
temp_name = ''
duplicate_count = 0

for column in dft.columns:
    # Get column index
    index = dft.columns.get_loc(column)

    # Apply filtering to remove unneccessary column text
    p = re.compile(r'<.*?>')
    format_col = p.sub('', column)
    format_col = format_col.replace(' ', '_')
    format_col = format_col.replace('.', '-')

    if format_col == temp_name:
        format_col = format_col + "_" + str(duplicate_count)
        duplicate_count += 1
    else:
        temp_name = format_col
        duplicate_count = 0
    
    # Apply rename
    dft_clean.rename(columns={dft_clean.columns[index]: format_col}, inplace=True)
```

```python
dft_clean.columns.values
```

```python
# Add 'Index' and 'Group' columns for usage in the framework solution
dft_clean['Index'] = np.arange(dft_clean.shape[0])
dft_clean['Group'] = 0
```

Export to .csv file

```python
dft_clean.to_csv('../data/2013_HHS_Delft_priva_16min_v2.csv')
```
